## Setup



```python
import os

from icecream import ic

os.chdir('../../datasets')
!ls

import wfdb

dset = 'St-Petersburg-INCART'
```

## wfdb



```python
rnm = 'I01'
fnm = f'{dset}/{rnm}'


```

```python
help(wfdb.Record)

sampto = 2 ** 14
rec = wfdb.rdrecord(fnm, sampto=sampto)
ic(rec.p_signal)
ic(rec.d_signal)
ic(rec.e_p_signal)
ic(rec.e_d_signal)
ic(rec.record_name)
ic(rec.n_sig)
ic(rec.fs)
ic(rec.counter_freq)
ic(rec.base_counter)
ic(rec.sig_len)
ic(rec.base_time)
ic(rec.file_name)
ic(rec.fmt)
ic(rec.samps_per_frame)
ic(rec.skew)
ic(rec.byte_offset)
ic(rec.adc_gain)
ic(rec.baseline)
ic(rec.units)
ic(rec.adc_res)
ic(rec.adc_zero)
ic(rec.init_value)
ic(rec.checksum)
ic(rec.block_size)
ic(rec.sig_name)
ic(rec.comments)

```

```python
help(wfdb.Annotation)

ann = wfdb.rdann(f'{dset}/I01', extension='atr', sampto=sampto)
ic(ann.record_name)
ic(ann.extension)
ic(ann.sample)
ic(ann.symbol)
ic(ann.subtype)
ic(ann.chan)
ic(ann.num)
ic(ann.aux_note)
ic(ann.fs)
ic(ann.label_store)
ic(ann.description)
ic(ann.custom_labels)
ic(ann.contained_labels)

```

```python
sampto = 2 ** 10

rec = wfdb.rdrecord(fnm, sampto=sampto)
ann = wfdb.rdann(f'{dset}/I01', extension='atr', sampto=sampto)
wfdb.plot_wfdb(record=rec, annotation=ann, title='Record', figsize=(12, 12), plot_sym=True, ecg_grids='all', time_units='seconds')

```
